This notebook is for creating plots from final results.

This loops over ratios, even though I'm only using one. I want to try to generalize the scripts so that I can recreate other ratios.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import corrections as cor
import uncertainties as unc
import iso_funcs as iso
```

```python
#Definitions for how this code will work

outf = "test/" #Folder to hold dataframes, so that different attempts can be compared, include '/' if not blank

ratios = [['He3','H2']]
```

These next 3 cells will plot Bin Centering, Coulomb, Radiative, and Isoscalar corrections

```python
for ratio in ratios:
    df = pd.read_csv(outf + "dataframes/final/" + ratio[0] + "_" + ratio[1] + ".csv", index_col=0)
    
    plt.figure(figsize=(16,9))
    plt.gca().set(ylim=(0.99,1.01), xlim=(0.1,0.9))
    plt.errorbar(df.BinCenter,df.BCCor,df.RelBCCError*df.BCCor,fmt='o',ms=10,mew=2,elinewidth=2)
    plt.title("Bin Centering Correction to " + ratio[0] + "/" + ratio[1],fontsize='xx-large')
    plt.xlabel('Bjorken $x$',fontsize='x-large')
    plt.ylabel("Bin Centering Correction factor",fontsize='x-large')
    plt.xticks(np.arange(0.1,0.901,0.1),fontsize='large')
    plt.yticks(fontsize='large')
    plt.grid(linewidth=3)
    plt.savefig("Plots/bcc.eps")
```
