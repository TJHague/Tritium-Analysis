```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
h3 = pd.read_excel("H3D2.xlsx")
he3 = pd.read_csv("He3D2.dat")
A3 = pd.read_csv("f2v.csv")
dp = pd.read_csv("dp.csv",index_col=0)
#he3.ratio *= 1.024
#h3.Final_Ratio *= 0.996
```

First without Normalization

```python
iso = pd.DataFrame(columns=['BinCenter','Iso'])
f2he3 = pd.DataFrame(columns=['BinCenter','F2r'])
f2h3 = pd.DataFrame(columns=['BinCenter','F2r'])
```

```python
iso.BinCenter = h3.Bin_Center
f2he3.BinCenter = h3.Bin_Center
f2h3.BinCenter = h3.Bin_Center
```

```python
iso.Iso = (h3.Final_Ratio + he3.ratio)/2
f2he3.F2r = (he3.ratio*3 - 2*2*iso.Iso)/((2-he3.ratio)*3 - 2*2)
f2h3.F2r = (h3.Final_Ratio*3 - 2*1*iso.Iso)/((2-h3.Final_Ratio)*3 - 2*1)
```

```python
plt.figure(figsize=(16,9))
plt.plot(f2he3.BinCenter,f2he3.F2r,"o",label="He3",ms=10,mew=2,fillstyle='none')
plt.plot(f2h3.BinCenter,f2h3.F2r,"o",label="H3",ms=10,mew=2,fillstyle='none')
plt.plot(A3.x,A3.f2r,"o",label="He3/H3",ms=10,mew=2,fillstyle='none')
plt.plot(dp.x,dp.np,"o",label="D/p KP",ms=10,mew=2,fillstyle='none')
#plt.plot(dp.x,dp.np_CJ,"o",label="D/p CJ",ms=10,mew=2,fillstyle='none')
plt.legend(fontsize='x-large')
plt.title("F2n/F2p Extraction - No Normalization",fontsize="xx-large")
```

```python
plt.figure(figsize=(16,9))
plt.plot(iso.BinCenter,iso.Iso,"o",fillstyle='none',ms=10,mew=2,label="A=3 Isoscalar Nucleus")
plt.plot(he3.x,he3.ratio,"o",fillstyle='none',ms=10,mew=2,label="He3/D2")
plt.plot(h3.Bin_Center,h3.Final_Ratio,"o",fillstyle='none',ms=10,mew=2,label="H3/D2")
plt.legend(fontsize='x-large')
plt.title("Yield Ratios - No Normalization",fontsize="xx-large")
```

Now add in Normalization

```python
he3.ratio *= 1.024
h3.Final_Ratio *= 0.996
```

```python
iso.Iso = (h3.Final_Ratio + he3.ratio)/2
f2he3.F2r = (he3.ratio*3 - 2*2*iso.Iso)/((2-he3.ratio)*3 - 2*2)
f2h3.F2r = (h3.Final_Ratio*3 - 2*1*iso.Iso)/((2-h3.Final_Ratio)*3 - 2*1)
```

```python
plt.figure(figsize=(16,9))
plt.plot(f2he3.BinCenter,f2he3.F2r,"o",label="He3",ms=10,mew=2,fillstyle='none')
plt.plot(f2h3.BinCenter,f2h3.F2r,"o",label="H3",ms=10,mew=2,fillstyle='none')
plt.plot(A3.x,A3.f2r,"o",label="He3/H3",ms=10,mew=2,fillstyle='none')
plt.plot(dp.x,dp.np,"o",label="D/p KP",ms=10,mew=2,fillstyle='none')
#plt.plot(dp.x,dp.np_CJ,"o",label="D/p CJ",ms=10,mew=2,fillstyle='none')
plt.legend(fontsize='x-large')
plt.title("F2n/F2p Extraction - With Normalization",fontsize="xx-large")
```

```python
plt.figure(figsize=(16,9))
plt.plot(iso.BinCenter,iso.Iso,"o",fillstyle='none',ms=10,mew=2,label="A=3 Isoscalar Nucleus")
plt.plot(he3.x,he3.ratio,"o",fillstyle='none',ms=10,mew=2,label="He3/D2")
plt.plot(h3.Bin_Center,h3.Final_Ratio,"o",fillstyle='none',ms=10,mew=2,label="H3/D2")
plt.legend(fontsize='x-large')
plt.title("Yield Ratios - With Normalization",fontsize="xx-large")
```
