```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
inf = 'test/'
norm = 1.024

emc = pd.read_csv(inf + "dataframes/final/He3_H2.csv")
# emc = pd.read_csv('emc_raw.dat')
# emcnorm = pd.read_csv('emc_norm.dat')
# KP = pd.read_csv('KP_He3_EMC.csv')
# raw = pd.read_csv('emc_noiso.dat')
#e3103 = pd.read_csv('E3103.csv')
emc['RatioNorm'] = emc.Ratio.apply(lambda x: x*norm)
emc['IsoNorm'] = emc.Isoscalar_Ratio.apply(lambda x: x*norm)
#raw['norm'] = raw.ratio.apply(lambda x: x*1.025)
```

Plots that are just my data

```python
plt.figure(figsize=(48,27))
plt.plot([0,1],[1,1],'k--',linewidth=5)
plt.errorbar(emc['BinCenter'],emc['Ratio'],emc['RelRatioError']*emc['Ratio'],fmt='C0s',label="$^3$He EMC Ratio",ms=30,mew=5,elinewidth=5)
#plt.errorbar(e3103['x'],e3103['Ratio']/e3103['Iso. corr.'],((e3103['Stat']**2)+(e3103['Sys']**2))**(1/2),fmt='o',fillstyle='none',label="E3103",ms=30,mew=5)
#plt.errorbar(emc['x'],emc['ratio'],emc['error'],fmt='o',fillstyle='none',label="Isoscalar $^3$He EMC Ratio",ms=30,mew=5,elinewidth=5)
#plt.plot(KP['x'],KP['ratio'],"b-",label="Kulagin-Petti Model Prediction",linewidth=10)
plt.gca().set(ylim=(0.9,1.2), xlim=(0.1,0.9))
plt.legend(loc=2,fontsize=60)
plt.title("Helium-3 EMC Ratio",fontsize=100)
plt.xlabel('Bjorken $x$',fontsize=80)
plt.ylabel(r'$\mathrm{\sigma_{^{3}He}/\sigma_{^{2}H}}$',fontsize=80)
plt.xticks(np.arange(0.1,0.901,0.1),fontsize=80,y=-0.02)
plt.yticks(np.arange(0.9,1.21,0.05),fontsize=80)
plt.grid(linewidth=3)
plt.text(0.15, 0.91,'PRELIMINARY',fontsize=300, color='gray', alpha=0.5)
plt.savefig('Plots/pres_pics/yield.eps')
plt.show()
```
